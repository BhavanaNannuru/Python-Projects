```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('image.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
inverted_image = 255 - gray_image

plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted Grayscale')
plt.axis('off')
plt.show()
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21),0)

plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
plt.show()
```

```python
inverted_blur = 255 - blurred

plt.imshow(inverted_blur, cmap='gray')
plt.title('Inverted Blurred Image')
plt.axis('off')
plt.show()
```

```python
# This function gives the pencil sketch effect
pencil_sketch = cv2.divide(gray_image, inverted_blur, scale=256.0)

plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()
```
